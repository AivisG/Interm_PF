```python
import matplotlib.pyplot as plt

# Example Data
x = [1, 2, 3, 4, 5]
y = [10, 15, 7, 20, 14]

# Create the Figure
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(x, y, marker="o", linestyle="-", label="Portfolio Balance")

# Add Title and Labels
ax.set_title("Portfolio Balance Over Time")
ax.set_xlabel("Time")
ax.set_ylabel("Balance")
ax.legend()

# Add Explanation Text Below the Plot
fig.text(0.1, -0.2, 
         "This plot shows the changes in portfolio balance over time.\n"
         "The line represents the balance after each transaction.\n"
         "Peaks indicate high profits, while dips may indicate drawdowns.\n"
         "Use this chart to analyze investment strategy performance.",
         fontsize=12, wrap=True)

# Adjust Layout to Avoid Overlap
plt.tight_layout()

# Show the Plot
plt.show()
```
